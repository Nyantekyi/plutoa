```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

```

```python
def geturldata(url):
    response = requests.get(url)
    response.raise_for_status()  # Check for request errors
    soup = BeautifulSoup(response.content, 'html.parser')
    df=pd.read_html(StringIO(response.text))[0]
    new_header = df.iloc[0] #grab the first row for the header
    df = df[1:] #take the data less the header row
    df.columns = new_header #set the header row as the df header
    return df

```

```python
import itertools

def generate_atc_codes():

    toplevel=['A','B','C','D','G','H','J','L','M','N','P','R','S','V']
    character = [chr(i) for i in range(ord('A'), ord('Z')+1)]
    number = [f'{i:02d}' for i in range(1, 100)]

    atc_codes = []
    
    for first, second, third, fourth in itertools.product(toplevel, number,character,character):
        atc_code = f'{first}{second}{third}{fourth}'
        atc_codes.append(atc_code)
    
    return atc_codes

# Generate and print the first 10 ATC codes as a preview
atc_codes = generate_atc_codes()
print(atc_codes[:10])  # Print first 10 ATC codes to check the output
```
